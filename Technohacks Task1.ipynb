# Importing Libraries

```python
import pandas as pd
import numpy as np
```

# Loading Data

```python
train_data = pd.read_csv("D:\\movies\\projects\\Technohacks\\train.csv")
test_data = pd.read_csv("D:\\movies\\projects\\Technohacks\\test.csv")
gs = pd.read_csv("D:\\movies\\projects\\Technohacks\\gender_submission.csv")
```

```python
# Combine train and test data for cleaning
combined_data = pd.concat([train_data, test_data], ignore_index=True)
```

# Data Cleaning

Cleaning of train data

```python
combined_data.isnull().sum()
```

```python
combined_data['Age'].fillna(combined_data['Age'].median(), inplace=True)
combined_data
```

```python
combined_data['Survived'].fillna(0, inplace=True)
combined_data
```

```python
combined_data['Embarked'] = combined_data['Embarked'].fillna(method = 'ffill')
combined_data
```

```python
#dropping unnecessary columns
combined_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
combined_data
```

```python
# Outlier Detection and Removal (using Z-score)
from scipy.stats import zscore
z_scores = zscore(combined_data['Fare'])
outliers = (np.abs(z_scores) > 3)
combined_data = combined_data[~outliers]
```

```python
# Save the cleaned dataset
combined_data.to_csv('cleaned_titanic_dataset.csv', index=False)

# Display a message indicating successful completion
print("Cleaning process completed. Cleaned dataset saved as 'cleaned_titanic_dataset.csv'.")
```
