Extracting Numerical and Categorical Columns

```python
numerical_cols = iris.select_dtypes(exclude='object').columns
print(numerical_cols)

categorical_cols = iris.select_dtypes(include='object').columns
print(categorical_cols)
```

```python
#Visualizing Numerical Columns using Hisplot
iris.hist(figsize=(10, 10), color='#00FFFF')
plt.show()
```

```python
#Visualizing Categorical columns using Barplot
for feature in categorical_cols:
    plt.figure(figsize=(5, 5))
    sns.countplot(x=feature, data=iris, palette='deep')
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```
